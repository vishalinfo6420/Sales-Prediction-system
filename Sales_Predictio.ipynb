```python
x = df[['TV', 'Radio', 'Newspaper']]
```

```python
x
```

```python
y = df['Sales']
```

```python
y
```

```python
print(x.head())
# print(y.head())
```

```python
print(y.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
x
```

```python
y_train
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
print(model.coef_)  ## coefficient 
```

```python
y_pred = model.predict(X_test)
```

```python
model.predict([[150, 25, 15]])
```

```python
# Compare Actual vs Predicted
result = pd.DataFrame({
    'Actual': y_test,
    'Predicted': y_pred
})

print(result.head())
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
r2 = r2_score(y_test, y_pred)
```

```python
print("R2 Score:", r2)
```

```python
# Mean Absolute Error (MAE)
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred)
```

```python
print("MAE", mae)
```

```python

# RMSE
from sklearn.metrics import mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("RMSE:", rmse)
```

```python
new_data = [[900, 45, 90]]  ##[TV,Radio,Newspaper]

sales_prediction = model.predict(new_data)

print("Predicted Sales:", sales_prediction[0])
```

```python
new_data = [[200, 40, 50]]
```

```python
sales_prediction = model.predict(new_data)

print("Predicted Sales:", sales_prediction[0])
```
